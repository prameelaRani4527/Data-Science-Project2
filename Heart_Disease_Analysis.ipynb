```python
import numpy as np #linear algebra
import pandas as pd #data processing
import matplotlib.pyplot as plt #data visualization
import seaborn as sns #data visualization
df=pd.read_csv("/content/drive/MyDrive/heart.csv")
df.head()
```

```python
#target distribution
df.target.value_counts()
```

```python
#show correlation matrix
corr=df.corr()
plt.figure(figsize=(18,10))
sns.heatmap(corr,annot=True)
plt.show()
```

```python
#plot target dist
#sns.ccountplot(df.target,hue='sex',palette=['green','red])
#sns.ccountplot(df.target,hue='sex',palette=['blue','yellow'])
sns.countplot(x=df.target,hue='sex',data=df,palette=['#1CA53B','red'])
#seaborn.pairplot(df,hue='size')
plt.xticks([0,1],labels=['not disease','disease'])
plt.title("Target Distribution")
plt.show()
```

```python
plt.figure(figsize=(18,10))
sns.countplot(x='age',hue='target',data=df,palette=['blue','yellow'])
plt.legend(["Haven't disease","Have disease"])
plt.title("Heart disease freuency for ages")
plt.xlabel('age')
plt.ylabel('frequency')
plt.show()
```

```python
plt.figure(figsize=(18,10))
sns.distplot(df.age[df['target']==0],bins=30,color='blue',label='not disease')
sns.distplot(df.age[df['target']==1],bins=30,color='yellow',label='disease')
plt.legend()
plt.title("Age distribution for heart disease")
plt.xlabel("age")
plt.ylabel("frequency")
plt.show()
```
